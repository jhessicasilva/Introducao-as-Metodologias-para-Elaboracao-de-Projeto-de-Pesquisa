# **Visualização de Dados com Python**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Carregando os Dados

O conjunto de dados escolhido para essa atividade foi 0 "*Diabetes Dataset - Diabetes Patients Data*", disponível no Kaggle. O objetivo do conjunto de dados é predizer, de forma diagnóstica, se um paciente tem diabetes, com base em certas medidas diagnósticas incluídas no conjunto de dados.

```python
dataset = pd.read_csv('diabetes.csv')
print(type(dataset))
dataset
```

**Descrição das colunas:**

* **Pregnancies**: número de gestações
* **Glucose**: nível de glicose no sangue
* **Bloodpressure**: medição da pressão arterial
* **SkinThickness**: espessura da pele
* **Insulin**: nível de insulina no sangue
* **BMI**: índice de massa corporal
* **DiabetesPedigreeFunction**: porcentagem de diabetes
* **Age**: idade 
* **Outcome**: 1-sim e 0-não

Comente aspectos que você achou interessantes do(s) conjunto(s) de dados escolhido(s) para cada tipo de visualização. O que esses gráficos te ajudaram a entender sobre os dados?




### Histograma
Os aspectos que achei interessante do conjunto de dados escolhido no tipo de visualização histograma foram:
* Foi possível observar, como nota-se em nossa sociedade, que a maioria das mulheres ou não tiveram gestações ou tiveram em média 3 gestações. Há casos em que algumas poucas mulheres gastaram mais de 6 vezes, mas o histograma mostra um reflexo da sociedade, em que esse é um caso que ocorre poucas vezes.
* Algumas variáveis apresentam medidas zeradas, como por exemplo, os histogramas de nível de glicose no sangue e o de índice de massa corporal.  Isso não significa que a pessoa não possui glicose no sangue ou que a pessoa não possui nenhuma massa corporal. Na realidade, isso deve significar, nesse caso, que talvez essas variáveis não foram medidas nas pacientes, então durante a construção do conjunto de dados, foi preferível colocar essa variável em zero, mostrando que o valor não foi colhido, visto que não faria sentido uma pessoa não possuir um nível de glicose no sangue, por exemplo.
 
 
### Gráfico de Barras
O aspectos que mais achei interessante do conjunto de dados escolhido no tipo de visualização de gráfico de barras, que foram feitos com relação ao dado se a mulher tem diabete ou não, foi o fato de que uma única variável não define se a pessoa tem diabete ou não, como pode ser observado principalmente nos gráficos de glicose, medição da pressão arterial, insulina, índice de massa corporal e porcentagem de diabetes.
 
 
### Gráfico de dispersão
Os aspectos que achei interessante do conjunto de dados escolhido no tipo de visualização de gráfico de dispersão, que foram feitos com relação à porcentagem de diabetes, foi que embora os gráficos apresentam as relações entre as variáveis do conjunto de dados com relação a porcentagem de diabetes no sangue, mostrando onde se encontram os principais valores médios, é possível verificar que esse conjunto de dados possui vários outliers, mostrando que há valores que fogem da normalidade.
 
-----------------
Esses gráficos me ajudaram a entender que, considerando o conjunto de dados utilizados, que não há correlações muito elevadas entre as variáveis do conjunto, mostrando que uma única variável (ex: glicose, insulina, índice de massa corporal, etc) não define se uma mulher possui ou não diabetes. No caso da diabete, é possível perceber, que o que define se uma pessoa possui a doença seria a combinação de várias condições/variáveis juntas. Por exemplo, duas pessoas diferentes com o mesmo índice de insulina não infere que ambas possuem diabete, pois enquanto uma pode ter a outra não visto que a uma pessoa pode ter cuidados mais saudáveis que a outra, menos glicose no sangue, menor índice de massa corporal, etc.
